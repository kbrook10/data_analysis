```python
# Step 11: Alter the legend of the plot
# .columns -> alters the legend titles

#bdata.columns = ['West', 'East']
bdata.resample('M').sum().plot()
```

```python
# Step 12: Create a rolling summation of the data to identify trends
# .rolling().sum() -> takes the sum of the rolling period indicated
# (Tip) - use the base unit of measure for the entries

#bdata.columns = ['West', 'East']
bdata.resample('D').sum().rolling(365).sum().plot()
```

```python
# Step 13: Alter the y axis limits to better view the data
# .rolling().sum() -> takes the sum of the rolling period indicated
# (Tip) - use the base unit of measure for the entries

#ax = bdata.resample('D').sum().rolling(365).sum().plot()
#ax.set_ylim(0,None)
```

```python
# Step 14: Create new column of the total data
# .rolling().sum() -> takes the sum of the rolling period indicated
# (Tip) - use the base unit of measure for the entries

#bdata['Total'] = bdata['West'] + bdata['East']
ax = bdata.resample('D').sum().rolling(365).sum().plot()
ax.set_ylim(0,None)
```

```python
# Step 15: Review the trend within individual days
# .groupby(date.index.time) -> groups by time of day
# .mean().plot() -> average the counts per day per time of day across the multiple years

bdata.groupby(bdata.index.time).mean().plot()
```

```python
# Step 16: Review the entire dataset using the pivot table
# .pivot_table() -> pass in the column value with the row and column headers (i.e. time of day and date)
# iloc -> request the index of a 5x5

pivoted = bdata.pivot_table('Total', index=bdata.index.time, columns=bdata.index.date)
pivoted.iloc[:5,:5]
```

```python
# Step 17: Plot the data of the pivot table
# this creates a line for each day of the yea
# legend = False -> removes the legend

pivoted.plot(legend=False, alpha=0.01)
```
